# Kuramoto Phase Synchronization Analysis

This notebook demonstrates how to load datalogger outputs, extract live oscillator phase data, compute the **Kuramoto Order Parameter** $r(t)$, and visualize phase alignment.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 1. Mathematical Formulation
The phase coherence of $N$ coupled oscillators is captured by the order parameter $r(t) \in [0, 1]$:

$$r(t) e^{i \psi(t)} = \frac{1}{N} \sum_{j=1}^N e^{i \theta_j(t)}$$

Where $\theta_j(t)$ is the phase angle of the $j$-th oscillator.

```python
def compute_order_parameter(phases):
    """Computes the Kuramoto order parameter for an array of phase angles (radians)."""
    N = len(phases)
    if N == 0:
        return 0.0
    complex_vectors = np.exp(1j * np.array(phases))
    return np.abs(np.sum(complex_vectors)) / N
```

## 2. Phase Synchronization Simulation
We model an active sweep from zero coupling (total chaos) to maximum coupling (total lock).

```python
steps = 100
N_oscillators = 8
time = np.linspace(0, 10, steps)

# Sweeping coupling factor from 0.0 to 1.0
coupling = np.linspace(0.0, 1.0, steps)
order_params = []

for c in coupling:
    # Standard noise variance scales inversely with coupling
    variance = (1.0 - c) * np.pi
    phases = np.random.normal(0, variance, N_oscillators)
    order_params.append(compute_order_parameter(phases))

df = pd.DataFrame({'coupling': coupling, 'order_parameter': order_params})
df.head()
```

## 3. Visualization
Plot the coherence index vs coupling level.

```python
plt.figure(figsize=(8, 5))
plt.plot(df['coupling'], df['order_parameter'], label='Order Parameter $r(t)$', color='#457b9d', linewidth=2.5)
plt.axvline(x=0.4, color='#e63946', linestyle='--', label='Critical Coupling Threshold ($K_c$)')
plt.xlabel('Coupling Coefficient ($K$)')
plt.ylabel('Phase Coherence ($r$)')
plt.title('Kuramoto Synchronization Phase Transition')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
