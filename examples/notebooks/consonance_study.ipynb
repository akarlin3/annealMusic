# Consonance & Dissonance Perception Analysis

This notebook demonstrates how to load, summarize, and analyze participant rating data collected using the **AnnealMusic Perceptual Experimentation Framework**.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## 1. Mock Data Ingestion
We load participant responses from `responses.csv` generated by our consonance experiment (comparing perfect fifths vs tritones).

```python
# Load or create mock response frame
mock_data = {
    'subject_id': [f'sub-{i:03d}' for i in range(1, 51)] * 2,
    'stimulus_id': ['perfect_fifth'] * 50 + ['tritone'] * 50,
    'response_value': list(np.random.randint(5, 8, size=50)) + list(np.random.randint(1, 4, size=50)),
    'rt_ms': list(np.random.normal(1200, 200, 50)) + list(np.random.normal(1500, 300, 50))
}
df = pd.DataFrame(mock_data)
df.head()
```

## 2. Statistical Analysis
Evaluate the mean rating differences and perform a paired T-test to measure statistical significance.

```python
fifth_ratings = df[df['stimulus_id'] == 'perfect_fifth']['response_value']
tritone_ratings = df[df['stimulus_id'] == 'tritone']['response_value']

print(f"Mean Perfect Fifth Rating: {fifth_ratings.mean():.2f} / 7.0")
print(f"Mean Tritone Rating: {tritone_ratings.mean():.2f} / 7.0")

# Perform paired T-test
t_stat, p_val = stats.ttest_rel(fifth_ratings, tritone_ratings)
print(f"T-statistic: {t_stat:.3f}, p-value: {p_val:.3e}")
```

## 3. Visualization
Render the visual comparisons.

```python
plt.figure(figsize=(8, 5))
plt.bar(['Perfect Fifth', 'Tritone'], [fifth_ratings.mean(), tritone_ratings.mean()], yerr=[fifth_ratings.std(), tritone_ratings.std()], capsize=10, color=['#2a9d8f', '#e76f51'])
plt.ylabel('Consonance Rating (1-7)')
plt.title('Perceived Consonance Comparison (N=50)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
